```python
source = donations[0:4999]

alt.Chart(source).mark_bar().encode(
    alt.X('Donation Amount:Q', bin=True),
    y='count()'
    # could also use alt.Y(aggregate='count', type='quantitative')
)
```

```python
#Consider looking at entire dataset donation amount
plt.hist(donations['Donation Amount'].dropna(), bins = 10000)
plt.xlabel('Donation Amt')
plt.xlim([0,500])
plt.ylabel('Probability')
plt.title('Histogram of donations')
plt.grid(True)
plt.show()
```

```python
#Look at some metadata
np.nanmean(donations['Donation Amount'])
np.nanstd(donations['Donation Amount'])
```

```python
df = pd.merge(donations, donors)
```

```python
df2 = df[0:4999]
alt.Chart(df2).mark_area(
    opacity=0.3,
    interpolate='step'
).encode(
    alt.X('Donation Amount', bin=alt.Bin(maxbins=100)),
    alt.Y('count()'),
    alt.Color(
        'Donor State')
    )
```
